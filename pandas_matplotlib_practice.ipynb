**Завдання 21–26:**
21. Створи DataFrame з кількома пропущеними значеннями (`NaN`) у числових і текстових стовпцях.  
22. Замінити всі `NaN` у текстових стовпцях на `"Unknown"`.  
23. Замінити `NaN` у числових стовпцях на середнє значення відповідного стовпця.  
24. Видалити всі рядки, де хоча б одне значення пропущене.  
25. Перевір кількість пропущених значень у кожному стовпці (`.isna().sum()`).  
26. Створи DataFrame з числовими і текстовими стовпцями та перевір `.dtypes`.  

````python
# Розв'язання для завдань 21–26
# Підказка: використовуй np.nan для створення пропусків.

# Твій код тут ↓
````

````python
# 21 Створи DataFrame з кількома пропущеними значеннями (NaN) у числових і текстових стовпцях.

df = pd.DataFrame({
    'Name':['Anna', 'NaN', 'Taras', 'Mark', 'Max'],
    'Level':['NaN', 'B2', 'A2', 'NaN', 'B1'],
    'Age':[34, 27, 30, 'NaN', 'NaN']
})

df
````

````python
df.info()
````

````python
df.describe()
````

````python
# 22 Замінити всі NaN у текстових стовпцях на "Unknown"

text_cols = df.select_dtypes(include='object').columns

df[text_cols] = df[text_cols].fillna("Unknown")

print(df)
````

````python
# 23 Замінити NaN у числових стовпцях на середнє значення відповідного стовпця

num_cols = df.select_dtypes(include='number').columns

for col in num_cols:
    mean_value = df[col].mean()
    df[col].fillna(mean_value, inplace=True)

print(df)
````

````python
# 24 Видалити всі рядки, де хоча б одне значення пропущене

df = pd.DataFrame({
    'Name':['Anna', 'NaN', 'Taras', 'Mark', 'Max'],
    'Level':['NaN', 'B2', 'A2', 'NaN', 'B1'],
    'Age':[34, 27, 30, 'NaN', 'NaN']
})

df = df.replace('NaN', np.nan)

df_clean = df.dropna()

print(df_clean)
````

````python
# 25 Перевір кількість пропущених значень у кожному стовпці (.isna().sum())

df = pd.DataFrame({
    'Name':['Anna', 'NaN', 'Taras', 'Mark', 'Max'],
    'Level':['NaN', 'B2', 'A2', 'NaN', 'B1'],
    'Age':[34, 27, 30, 'NaN', 'NaN']
})

df = df.replace('NaN', np.nan)

print(df.isna().sum())
````

````python
# 26 Створи DataFrame з числовими і текстовими стовпцями та перевір .dtypes

df.dtypes
````

````python
# 28 Знайди загальний дохід (Price * Quantity) і додай його як новий стовпець "Revenue"

df_sale['Revenue'] = df_sale['Price'] * df_sale['Quantity']

df_sale
````

````python
# 34 Відсортуй категорії за середнім доходом

df_sale.groupby("Category")["Revenue"].mean().sort_values(ascending=False)
````

````python
# 36a Створи два DataFrame: customers і orders:
#    - customers: CustomerID, Name, City  
#    - orders: OrderID, CustomerID, Amount  
# 36b З’єднай їх за CustomerID (pd.merge)

df_customers = pd.DataFrame({
    'CustomerID':[11, 26, 29, 4, 12, 27, 13]
    , 'Name':['Yuliia', 'Dmytro', 'Petro', 'Valentyna', 'Mykola', 'Lidiya','Svitlana']
    , 'City':['Poltava', 'Lviv', 'Chernovtsy', 'Kherson', 'Stanislav', 'Megion', 'Stanislav']
})

df_customers
````

````python
df_customers.info()
````

````python
df_customers.describe()
````

````python
df_customers.shape
````

````python
df_customers.sample()
````

````python
df_orders = pd.DataFrame({
    'OrderID':[5, 9, 46, 54, 76, 84, 74]
    , 'CustomerID':[11, 26, 29, 4, 12, 27, 13]
    , 'Amount':[20, 16, 56, 71, 49, 41, 51]
})

df_orders
````

````python
df_orders.info()
````

````python
df_orders.describe()
````

````python
# 36b З’єднай їх за CustomerID (pd.merge)

pd.merge(df_customers, df_orders, on='CustomerID')
````

````python
# 37 Об’єднай дві таблиці вертикально (pd.concat)

df_customers_orders = pd.concat([df_customers, df_orders], ignore_index=True)

print(df_customers_orders)
````

````python
# 38 Знайди клієнтів, які **не мають** замовлень (left join + NaN у "Amount")

df_customers_orders_1 = df_customers.merge(df_orders, on='CustomerID', how='left')

print(df_customers_orders_1)
````

````python
# 39 Збережи результат у CSV-файл (.to_csv("customers_orders.csv", index=False)) і потім зчитай його назад (pd.read_csv)

df_customers_orders_1.to_csv("customers_orders.csv", index=False)

df_loaded = pd.read_csv("customers_orders.csv")
````

````python
# 40 Перевір, чи зчитані типи даних збігаються з оригіналом (порівняй .dtypes)

print(df_customers_orders_1.dtypes) 
print(df_loaded.dtypes)
````

````python
# 41 Створи DataFrame з колонкою "date" від "2024-01-01" до "2024-01-10" (використай pd.date_range)

dates = pd.date_range(start="2024-01-01", end="2024-01-10")
df_decade_1_jan = pd.DataFrame({"date": dates})

print(df_decade_1_jan)
````

````python
# 42 Додай новий стовпець "day_of_week", що показує день тижня

df_decade_1_jan["date"] = pd.to_datetime(df_decade_1_jan["date"])
df_decade_1_jan["day_of_week"] = df_decade_1_jan["date"].dt.day_name()

print(df_decade_1_jan)
````

````python
# 43 Вибери всі дні вихідного (Saturday, Sunday)

weekends = df_decade_1_jan[df_decade_1_jan["day_of_week"].isin(["Satureday", "Sunday"])]

print(weekends)
````

````python
# 44 Знайди різницю між найпізнішою та найранішою датою (date.max() - date.min())

difference = df_decade_1_jan["date"].max() - df_decade_1_jan["date"].min()

print(difference)
````

````python
# 45 Додай колонку "month" і порахуй, скільки записів у кожному місяці

df_decade_1_jan["month"]= df_decade_1_jan["date"].dt.month
df_decade_1_jan["month"].value_counts()
````
